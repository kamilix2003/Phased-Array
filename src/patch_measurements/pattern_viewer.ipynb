```python
import numpy as np
import matplotlib.pyplot as plt
import pattern_import

import ipywidgets as widgets
import IPython.display as display
```

```python
base_path = "C:/Users/kamaj/Code/Phased-Array/src/patch_measurements/"
file_path_1 = base_path + "antena_v01_azymut.txt"
file_path_2 = base_path + "antena_v01_azymut90deg_5xoffset.txt"
pattern1 = pattern_import.RadiationPattern(file_path_1)
pattern2 = pattern_import.RadiationPattern(file_path_2)
```

```python
pattern = pattern2
slider_list = []

azimuth_slider = widgets.FloatRangeSlider(value = (pattern.azimuth[0], pattern.azimuth[-1]),
                         min = pattern.azimuth[0],
                         max = pattern.azimuth[-1],
                         step = pattern.config['Azimuth step [deg]'],
                         description = 'Azimuth Range:',
                         continuous_update = True)
slider_list.append(azimuth_slider)

elevation_slider = widgets.FloatSlider(value = (pattern.elevation[0] + pattern.elevation[-1]) / 2,
                         min = pattern.elevation[0],
                         max = pattern.elevation[-1],
                         step = pattern.config['Elevation step [deg]'],
                         description = 'Elevation Range:',
                         continuous_update = True)
slider_list.append(elevation_slider)

if pattern.polarization is not None:
  polarization_slider = widgets.FloatSlider(value = (pattern.polarization[0] + pattern.polarization[-1]) / 2,
                                                min = pattern.polarization[0],
                                                max = pattern.polarization[-1],
                                                step = pattern.config['Polarization step [deg]'],
                                                description = 'Polarization Range:',
                                                continuous_update = True)
  slider_list.append(polarization_slider)
  
if pattern.offset is not None:
  offset_slider = widgets.FloatRangeSlider(value = (pattern.offset[0], pattern.offset[-1]),
                          min = pattern.offset[0],
                          max = pattern.offset[-1],
                          step = pattern.config['Offset step [mm]'],
                          description = 'Offset Range:',
                          continuous_update = True)
  slider_list.append(offset_slider)

frequency_slider = widgets.FloatSlider(value = (pattern.frequency[0] + pattern.frequency[-1]) / 2,
                          min = pattern.frequency[0],
                          max = pattern.frequency[-1],
                          step = 0.01,
                          description = 'Frequency Range:',
                          continuous_update = True)
slider_list.append(frequency_slider)

def update_plot(change):
  with output:
    output.clear_output(wait=True)
    data, ranges = pattern.get_pattern(frequency=np.array([frequency_slider.value]))
    current_pattern = data[:, 0, 0, pattern.get_value_range_idx('offset', offset_slider.value), 0]# - np.max(data)
    labels = [f"Offset: {offset:.2f} mm" for offset in pattern.offset[pattern.get_value_range_idx('offset', offset_slider.value)]]
    plt.figure(figsize=(8, 5))
    plt.polar(np.deg2rad(pattern.azimuth[ranges["azimuth"]]), current_pattern, label=labels)
    plt.legend()
    plt.title("Radiation Pattern (Azimuth Cut)")
    plt.grid()
    plt.show()

for slider in slider_list:
  slider.observe(update_plot, names='value')

sliders = widgets.VBox(slider_list)
output = widgets.Output()
display.display(sliders, output)
                            
```
